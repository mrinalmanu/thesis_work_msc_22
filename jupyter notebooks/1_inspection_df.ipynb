# mutating datasets

```python
mcr_df_hr_n["id"] = mcr_df_hr_n.index
mcr_df_gr_n["id"] = mcr_df_gr_n.index
```

```python
ncol_dg = ['id', 'Progress', 'UserLanguage', 'Collection']
```

```python
dfx = pd.melt(mcr_df_hr_n, id_vars=list(ncol_dg))
dfx["Variable"] = dfx["variable"]
del dfx["variable"]
dfx
```

```python
efx = pd.melt(mcr_df_gr_n, id_vars=list(ncol_dg))
efx["Variable"] = efx["variable"]
del efx["variable"]
efx
```

```python
t_hr = pd.merge(dfx, hr, on='Variable')
t_hr
```
